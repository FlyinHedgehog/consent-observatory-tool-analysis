```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (14, 8)

# CONFIG: Set dataset name here (corresponds to folder in analysis_output/)
DATASET_NAME = 'tranco_QWJY4'

print('✓ Imports ready')
```

## Load Analysis Results

```python
# Load Excel files from analysis folder
data_dir = Path('../analysis_output') / DATASET_NAME

if not data_dir.exists():
    print(f'✗ Error: {data_dir} not found!')
    print(f'✓ Available datasets:')
    analysis_root = Path('analysis_output')
    if analysis_root.exists():
        for d in analysis_root.iterdir():
            if d.is_dir():
                print(f'  - {d.name}')
    else:
        print('  analysis_output folder not found')
else:
    try:
        cookies_df = pd.read_excel(data_dir / 'cookies.xlsx')
        buttons_df = pd.read_excel(data_dir / 'buttons.xlsx')
        sites_summary_df = pd.read_excel(data_dir / 'sites_summary.xlsx')
        
        print(f'✓ Loaded analysis results for: {DATASET_NAME}')
        print(f'  - {len(sites_summary_df)} sites analyzed')
        print(f'  - {len(cookies_df)} cookies found')
        print(f'  - {len(buttons_df)} buttons detected')
    except FileNotFoundError as e:
        print(f'✗ Error loading files: {e}')
        print(f'✓ Files in {data_dir}:')
        for f in data_dir.glob('*'):
            print(f'  - {f.name}')
```
